```python
import gradio as gr

def create_object(arg):
    return gr.Textbox(value=arg)

my_objects = {}

test_list =["first", "second"] 
for i in test_list:
    object_name = f"object_{i}"
    new_object = create_object(i)
    my_objects[object_name] = new_object

# Accessing objects by their assigned names
first_object = my_objects["object_first"]
second_object = my_objects["object_second"]

with gr.Blocks() as test:
    with gr.Row():
        first_object.render()
    with gr.Row():
        second_object.render()

test.launch()
```
